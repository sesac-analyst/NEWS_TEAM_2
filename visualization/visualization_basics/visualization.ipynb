```python
import pandas as pd
```

```python
# Frequency_naver 데이터
Frequency = pd.read_csv("./Frequency_naver_3months.csv")
print(len(Frequency))
Frequency=Frequency[Frequency['frequency']>1]
print(len(Frequency))
# 등장횟수별 정렬
Frequency=Frequency.sort_values('frequency',ascending=False)

# 인덱스 재설정
df = Frequency.reset_index(drop=True)
```

```python
# barplot

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프 크기 설정
plt.figure(figsize=(10, 10))

# 시각화: 모든 단어의 빈도수 막대 그래프 (상위 500개 데이터 사용)
sns.barplot(x='word', y='frequency', data=df[0:5], palette='viridis')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=30)
plt.xlabel('Word', fontsize=20)
plt.ylabel('Frequency', fontsize=20)

# y축 범위 설정
plt.ylim(2500,4600)
plt.xticks(fontsize=30)

# 그래프 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('barplot5_3months.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프 크기 설정
plt.figure(figsize=(20,5))

# 시각화: 모든 단어의 빈도수 막대 그래프
sns.barplot(x='word', y='frequency', data=df[0:1001], palette='viridis')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=50)
plt.xlabel('Frequency', fontsize=30)
plt.ylabel('Word', fontsize=30)

# y축 범위 설정 (최대값을 1500으로 설정)
plt.ylim(0, 1000)

# X축에 100, 200, 300, 400, 500 눈금 표시
length=[]
for i in range(11):
    length.append(i*100)
plt.xticks(length, length)

# X축과 Y축 눈금 글씨 크기 설정
plt.xticks(fontsize=20)  # X축 눈금 글씨 크기 설정
plt.yticks(fontsize=20)  # Y축 눈금 글씨 크기 설정

# 그래프 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('barplot_3months.png')
plt.show()

```

```python
from wordcloud import WordCloud

# 한글 폰트 경로 설정 (Windows에서 맑은 고딕 폰트 경로)
font_path = "C:/Windows/Fonts/malgun.ttf"

# 워드 클라우드 생성
word_freq = dict(zip(df['word'], df['frequency']))
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축을 없애기

# 그래프 출력
plt.tight_layout()
plt.savefig('WordCloud_3months.png')
plt.show()
```

```python
# 빈도 분포 확인 (KDE Plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(df['frequency'], shade=True, color='blue')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=14)
plt.xlabel('Frequency', fontsize=12)

# 그래프 출력
plt.tight_layout()

plt.savefig('KDE Plot.png')
plt.show()
```

```python
import seaborn as sns

# 상자 그림으로 데이터 분포 확인
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['frequency'], color='lightblue')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=14)
plt.xlabel('Frequency', fontsize=12)

# 그래프 출력
plt.tight_layout()
plt.savefig('boxplot.png')
plt.show()
```
