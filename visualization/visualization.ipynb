```python
# Frequency_naver 데이터
Frequency = pd.read_csv("../tokenized/Frequency_naver.csv")
Frequency=Frequency[Frequency['frequency']>1]

# 'frequency' 열의 글자 수가 2개 이상인 행만 필터링
df = Frequency[Frequency['word'].astype(str).str.len() >= 2]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프 크기 설정
plt.figure(figsize=(100,100))

# 시각화: 모든 단어의 빈도수 막대 그래프
sns.barplot(x='word', y='frequency', data=df, palette='viridis')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=14)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Word', fontsize=12)

# y축 범위 설정 (최대값을 1500으로 설정)
plt.ylim(0, 1500)

# 그래프 레이아웃 조정 및 출력
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

# 한글 폰트 경로 설정 (Windows에서 맑은 고딕 폰트 경로)
font_path = "C:/Windows/Fonts/malgun.ttf"

# 워드 클라우드 생성
word_freq = dict(zip(df['word'], df['frequency']))
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축을 없애기

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
# 빈도 분포 확인 (KDE Plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(df['frequency'], shade=True, color='blue')

# 그래프 제목 및 라벨 설정
plt.title('Keyword Frequencies in News Data', fontsize=14)
plt.xlabel('Frequency', fontsize=12)

# 그래프 출력
plt.tight_layout()
plt.show()
```
